```python
import numpy as np
#import cv2, pydicom
import os
import pydicom
import matplotlib.pyplot as plt
from pydicom.pixel_data_handlers.util import apply_modality_lut, apply_voi_lut, apply_windowing
import shutil
```

```python
window_center = 50
window_width = 350
```

```python
path = '/home/ncp/workspace/202002n035/035.신장암 진단을 위한 의료 영상 데이터/01.데이터/신장암2/1.Training/원천데이터/'
```

```python
# if re-creation needed
shutil.rmtree('/home/ncp/workspace/blocks1/kidneyData_windowing/TRAIN/')
shutil.rmtree('/home/ncp/workspace/blocks1/kidneyData_windowing/TEST/')
```

```python
# make dir
# kidneyData_windowing 폴더는 만들고 돌려야 함

def makedir(path): 
    if not os.path.exists(path): 
        os.mkdir(path)
        
dir = '/home/ncp/workspace/blocks1/kidneyData_windowing/'
makedir(os.path.join(dir, "TRAIN"))
makedir(os.path.join(dir, "TEST"))

makedir(os.path.join(dir, "TRAIN", "ART"))
makedir(os.path.join(dir, "TEST", "ART"))

makedir(os.path.join(dir, "TRAIN", "PRE"))
makedir(os.path.join(dir, "TEST", "PRE"))
```

```python
#Except patient no. 108, 148, 184, 309

for f in range(1,108): 
    if f < 10:
        folder = '02_01_000{}'.format(f)
    elif f < 100:
        folder = '02_01_00{}'.format(f)
    else:
        folder = '02_01_0{}'.format(f)
    path_ = path + folder 
    dicom_path = path_+'/'+folder+'_ART/'
    file_names = os.listdir(dicom_path)
    #print(file_names) #['0002.dcm', '0004.dcm', 
    #print(folder)#02_01_0001
    for name in file_names:
        #print(folder+"_"+name+'.png') #02_01_0006_0062.DCM.png
        #print(dicom_path+name)#full
        
        slice = pydicom.read_file(dicom_path+name)
        s = int(slice.RescaleSlope) #slope
        b = int(slice.RescaleIntercept) #intercept
        ct_image = s * slice.pixel_array + b
    
        slice.WindowCenter = window_center
        slice.WindowWidth = window_width
    
        w_img = apply_windowing(ct_image, slice)
        plt.imsave('/home/ncp/workspace/blocks1/kidneyData_windowing/TRAIN/ART/'+folder+"_"+name[0:4]+'.png', w_img,  cmap='gray')
```
