```python
final = pd.read_csv('MME_final.csv', encoding = 'euc-kr', low_memory = False)
final
```

```python
tmp = final.copy()
tmp
```

```python
col = tmp.columns
```

```python
# 55555 ~ 99999 전부 NaN으로 대체
tmp = tmp[tmp[col] != 55555]

tmp = tmp[tmp[col] != 66666]

tmp = tmp[tmp[col] != 77777]

tmp = tmp[tmp[col] != 88888]

tmp = tmp[tmp[col] != 99999]
```

```python
tmp
```

```python
weight_cycle_col = ['기수', 'NIHID', 'EDATE', 'SEX', 'VISITALL', 'AGE', 'WEIGHT', 'HEIGHT', 
                   'WAIST', 'BODYFAT', 'TCHL_ORI', 'SMOKE', 'SBP_L', 'SBP_R', 
                   'CREATININ_TR', 'CREATININ_ORI']
tmp = tmp[weight_cycle_col]
tmp
```

```python
tmp['CREA_ADJUSTED'] = 0.0

for idx in range(len(tmp)): 
    if tmp['기수'][idx] == 'A01' or tmp['기수'][idx] == 'A06' or tmp['기수'][idx] == 'A07': 
        tmp['CREA_ADJUSTED'][idx] = tmp['CREATININ_TR'][idx]
    else: 
        tmp['CREA_ADJUSTED'][idx] = tmp['CREATININ_ORI'][idx]    
```

```python
tmp.isna().sum()
```

```python
# CREA_ADJUSTED 계산했으니 CREATININ_TR, CREATININ_ORI 열 drop
tmp.drop(['CREATININ_TR', 'CREATININ_ORI'], axis = 1, inplace = True)
tmp
```

```python
# tmp = 전체 70210개 중 쓸 칼럼만 뽑은 df

tmp2 = tmp.copy()
```

```python
# tmp2 = tmp 전체에서 null값 다 없앤 df

tmp2 = tmp2.dropna().reset_index(drop = True)
tmp2
```

```python
# onehot encoding

dummy_cols = ['SMOKE', 'SEX']
tmp2 = pd.get_dummies(tmp2, columns = dummy_cols)
tmp2
```

```python
# SBP, DBP 평균값으로 대체 후 L, R열 drop

tmp2['SBP'] = (tmp2['SBP_L'] + tmp2['SBP_R']) / 2.0
tmp2.drop(['SBP_L', 'SBP_R'], axis = 1, inplace = True)
```

```python
tmp2
```

```python
# BODYFAT 단위 맞추기

tmp2['BODYFAT'] = tmp2['BODYFAT'] * 1000
tmp2
```

```python
# BODYFAT 0 이하인 사람 자르기

tmp2 = tmp2.loc[tmp2['BODYFAT'] >= 0].reset_index(drop = True)
```

```python
tmp2
```

```python
tmp2
```

```python
tmp2['WEIGHT'] = tmp2['WEIGHT'] * 1000
tmp2
```

```python
tmp2['LBM'] = tmp2['WEIGHT'] - tmp2['BODYFAT']
tmp2
```

```python
tmp2.to_csv('MME_preprocessed_eGFR.csv', encoding = 'euc-kr', index = False)
```
