### 2021.09.14 ~ 2021.03.01 (완료)

```python
# 일별 확진자 수 테이블 생성
# link index (5918, 4546), 0시 기준' 이 title에 있을 때  --- 0916 여기서부터 돌리기

data = pd.DataFrame()
for idx in range(5918, 4546, -1):   # 2021.09.14 ~ 2021.03.01
    
    k = pd.DataFrame()  # 초기화
    response = requests.get('http://ncov.mohw.go.kr/tcmBoardView.do?brdId=3&brdGubun=31&dataGubun=&ncvContSeq={}&contSeq={}&board_id=312&gubun=ALL'.format(idx, idx))
    html = response.text
    soup = BeautifulSoup(html, 'html.parser')
    
    # root = lxml.html.fromstring(response.content)

    # 제목
    # title = soup.select('#content > div > div.board_view > div.bv_content > div.bvc_txt > p > p:nth-child(1)')  # 제목 (인덱스용)
    #vaccine = soup.select('#content > div > div.board_view > div.bv_content > div > p > table:nth-child(10)')
    #print(vaccine)
    
    title = soup.find('h4')  ## 찐 제목
    title = str(title)
    print(title.find('0시 기준'))
    
    if ('0시 기준' in title) and ('국내 발생 및 예방접종 현황' in title):   ## '코로나19 국내 발생 및 예방접종 현황(9.14., 0시 기준)' 제목형식과 일치할때
        print(title)
        # 국내 발생 확진자 현황 테이블 가져오기
        
        lst = soup.select('table')

        if len(lst) != 0:    # 테이블이 하나라도 존재할때
            for i in range(len(lst)):   # 테이블 개수만큼 검사
                tmp = str(lst[i])    # i번째 테이블 담기
                table_df_list = pd.read_html(tmp)
                table_df_tmp = table_df_list[0]  
        
                if (len(table_df_tmp) == 3) and (len(table_df_tmp.columns) > 10): # 행 개수 3개, 칼럼 개수 10개 이상 만족하는 테이블 확진자밖에 없음
                    table_df = table_df_tmp
        
        k = table_df.transpose()
        k = k.rename(columns = k.iloc[0])
        #print(k)
        k.drop([0], axis = 0, inplace = True)

        colname1 = title[26:31] + ' ' + '신규'
        colname2 = title[26:31] + ' ' + '누계'

        k.rename(columns = {'신규' : colname1, '누계' : colname2}, inplace = True)
        # k.drop(['구분'], axis = 1, inplace = True)
        
        data = pd.concat([data, k], axis = 1)
   
        print(response.url)
```
