```python
tmpin2 = tmpin2.drop(['DT', '시군구', 'AGE', 'SEX_TYPE'], axis = 1)
```

```python
em = pd.DataFrame(test.groupby(['DT', '시군구', 'AGE', 'SEX_TYPE']).sum()['ATOPIC_em_total']).reset_index()
```

```python
em = em.drop(['DT', '시군구', 'AGE', 'SEX_TYPE'], axis = 1)
```

```python
SUM = pd.DataFrame(test.groupby(['DT', '시군구', 'AGE', 'SEX_TYPE']).sum()['ATOPIC_SUM']).reset_index()
```
