```python
tmp = []
rows = []

for item in data:
        confcase=item.find("confcase")
        
        confcaserate=item.find("confcaserate") 
        createdt=item.find("createdt") 
        criticalrate=item.find("criticalrate")
        
        death=item.find("death") 
        
        deathrate=item.find("deathrate") 
        
        gubun=item.find("gubun") 
    
        seq=item.find("seq") 
        
        updatedt=item.find("updatedt") 
    
        rows = [confcase.get_text(),confcaserate.get_text(), 
             createdt.get_text(),criticalrate.get_text(), 
              death.get_text(), deathrate.get_text(),  
              gubun.get_text(), seq.get_text(),
              updatedt.get_text()]
        
        tmp.append(rows)
```

```python
final=pd.DataFrame(tmp)
final
```

```python
final.columns = ["confCase","confCaseRate", 
             "createDt","criticalRate", 
              "death", "deathRate",  
              "gubun", "seq",
              "updateDt"]
```

```python
from bs4 import BeautifulSoup
import pandas as pd
soup=BeautifulSoup(request.content,'html.parser')
data=soup.find_all('item') #data찍어서 항목확인
```

```python
#필요한 변수 추출
tmp = []
rows = []

for item in data:
        createDt=item.find("createdt")
        deathCnt=item.find("deathcnt") 
        defCnt=item.find("defcnt") 
        gubun=item.find("gubun")
        gubunCn=item.find("gubuncn") 
        gubunEn=item.find("gubunen") 
        incDec=item.find("incdec") 
    
        isolClearCnt=item.find("isolclearcnt") 
        isolIngCnt=item.find("isolingcnt") 
        localOccCnt=item.find("localocccnt") 
    
        overFlowCnt=item.find("overflowcnt") 
        qurRate=item.find("qurrate") 
        seq=item.find("seq") 
        stdDay=item.find("stdday") 
        updateDt=item.find("updatedt") 
    
        rows = [createDt.get_text(),gubun.get_text(), 
             deathCnt.get_text(),defCnt.get_text(), 
              gubunCn.get_text(), gubunEn.get_text(), 
              incDec.get_text(), isolClearCnt.get_text(),
              isolIngCnt.get_text(), localOccCnt.get_text(),
              overFlowCnt.get_text(),qurRate.get_text(),
              seq.get_text(), stdDay.get_text(),
              updateDt.get_text()]
        
        tmp.append(rows)
```

```python
final=pd.DataFrame(tmp)
final.head()
```

```python
final.columns = ["createdt", "gubun", "deathCnt", "defCnt", "gubunCn", "gubunEn", "incDec", "isolClearCnt", "isolIngCnt", "localOccCnt", 
                 "overFlowCnt", "qurRate", "seq", "stdDay", "updateDt"]
```
